## Categorical Plots

```python
"""
barplot
countplot
boxplot/ whiskerplot
violinplot
strippplot
swarmplot
catplot

In categorical plots, we will see the distributions of the categorical columns.
"""
```

```python
#1. barplot
#It is the general plot which is used to aggregate the categorical data based off of some functions and by default that's the mean.
#We can consider this as a visualization of a groupby() method.
```

```python
df.head()
```

```python
df.query('sex == "Male"').mean(numeric_only=True)
```

```python
#we can see the mean of total_bill is 20.744076.
```

```python
sns.barplot(x='sex',y='total_bill',data=df)  # Here by default plotting label of total_bill is based on mean of all datas.
# From this plot we can infer that the mean of total_bill for the males is higher than that of females. 
```

```python
# we can aggregate the barplot based on other estimators like mean, median, std and so on.
sns.barplot(x='sex', y='total_bill', data=df, estimator= np.median)
```

```python
#2. countplot
#It is similar to bar plot except that the estimator is explicitly counting the number of occurences
```

```python
df.query('sex == "Male"').count()
```

```python
df.query('sex == "Female"').count()[0]
```

```python
sns.countplot(x='sex', data=df, palette= 'Set2')
# from this plot we can see that we can see that the y axis is having the count values and x axis a categorical column
# This is also a visualisation of df['gender'].count()
```

```python
#3. boxplot / whiskerplot
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='Set1') 
```

## Grids

```python
#GRIDS
#Here we will be using grids ability to automate subplots.
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()   
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
```

```python
g.map(sns.scatterplot)
# If we want to specify a certain kind of plots for diagonal subplots like in pairplot
g.map_diag(plt.hist)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

## Regression Plots

```python
#Regression plots are genrally used in Machine learning algorithm called linear regression.
#In this we can see that the seaborn library actually calling the matplotlib under the hood.
```

```python
#1. lmplot
#It allows us to display linear models with seaborn.
```

```python
df=sns.load_dataset("tips")
```

```python
sns.lmplot(x='total_bill',y='tip',data=df) 
#it is same as the jointplot with kind="reg".
```

```python
sns.lmplot(x='total_bill',y='tip',data=df, hue='sex') 
# this tells us that the total_bill vs tips linear model is almost same for both male and female.
```

```python
sns.lmplot(x='total_bill',y='tip',data=df, hue='sex', markers=['o','x'], palette='Set1')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df, col='sex')   
# instead of differentiating with 2 different colors we can plot two different plots.
```

```python
# using this concept we can make similar to Facetgrid
sns.lmplot(x='total_bill',y='tip',data=df, col='sex', row= 'time') 
```

```python
sns.lmplot(x='total_bill',y='tip',data=df, col='sex', row= 'time', hue= 'smoker') 
```

```python
# To change the aspect ratio and size we use aspect and size parameter respectively
sns.lmplot(x='total_bill',y='tip',data=df, col='sex', row= 'time', hue= 'smoker', aspect=1.5, height=4) 
```
